```python
import os
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

connection=mysql.connector.connect(host='sql', user='sloth',passwd='sloth',auth_plugin='mysql_native_password', database='database')

reportes=pd.read_sql_query('SELECT * FROM database.reportes',connection)

fig_dims = (16, 10)
fig, ax = plt.subplots(figsize=fig_dims)

sns.barplot(x=reportes['MES'], y=reportes['BAJO_INFLUENCIA'], ax=ax)
plt.title('Agresores bajo la influencia por mes')
plt.xlabel('Meses')
plt.ylabel('Número de agresores')
plt.show()
```

```python
fig_dims = (16, 5)
fig, ax = plt.subplots(figsize=fig_dims)

sns.scatterplot(x=reportes['MES'], y=reportes['HUBO_MUERTO'], ax=ax, s=100)
plt.title('Casos con muertos por mes')
plt.xlabel('Meses')
plt.ylabel('Casos con muertos')
plt.show()
```

```python
fig_dims = (16, 5)
fig, ax = plt.subplots(figsize=fig_dims)
plt.plot(reportes['MES'], reportes['USO_DE_ARMAS'], color = 'r')
plt.title('Casos con armas por mes')
plt.xlabel('Meses')
plt.ylabel('Casos con armas')
plt.show()
```
